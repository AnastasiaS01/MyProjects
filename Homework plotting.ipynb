```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
population = pd.read_csv('data/population_by_country_2020.csv')
```

```python
c = population.sort_values(by='Population (2020)', ascending=False)[['Country (or dependency)', 'Population (2020)']].head(5)
```

```python
c = population.sort_values(by='Population (2020)', ascending=False)[['Country (or dependency)', 'Population (2020)']].head(5)
colors = ['#d53032', '#ff9218', '#000080', 'purple', '#006633']
plt.figure(figsize=(10, 6))
plt.barh(c['Country (or dependency)'], c['Population (2020)'], color=colors)
plt.xlabel('Population (millioms)')
plt.ylabel('Country (or dependency)')
plt.title('TOP 5 population countries 2020')

plt.gca().invert_yaxis()  
plt.show()
```
