Create 5 different visualization with your insights.

```python
#Visualisation 1: Number of flights according to the hour of day for different statuses
#It can help for airport staff to understand the busiest hours 
#and the distribution of different statuses during the day.

statuses = ['DEPARTED', 'CANCELED', 'DELAYED'] 
filtered_df = final_upgrade[final_upgrade['Status'].isin(statuses)]
plt.figure(figsize=(12, 6))
colors = {'DEPARTED': 'green','CANCELED': 'red','DELAYED': 'yellow'}
filtered_df['hour'] = filtered_df['PlanTime'].dt.hour
grouped = filtered_df.groupby(['hour', 'Status']).size().reset_index(name='counts')

sb.barplot(data=grouped, x='hour', y='counts', hue='Status', palette=colors, dodge=True)
plt.title('Number of flights according to the hour of day for different statuses')
plt.xlabel('Hour of day')
plt.ylabel('Number of flights')
plt.legend(title='Status')  
plt.xticks(range(0, 24, 2)) 
plt.show()
```

```python
#Visualization 2: Delays according to the number of flights in week
#Knowing this can help airlines and airports inform passengers in advance about possible delays, 
#increase the number of staff on these days to reduce delays and possible losses

final_upgrade['Total Delay (minutes)'] = (final_upgrade['ActualTime'] - final_upgrade['PlanTime']).dt.total_seconds() / 60
final_upgrade['DayOfWeek'] = final_upgrade['PlanTime'].dt.dayofweek
final_upgrade['DayOfWeek'] = final_upgrade['DayOfWeek'].replace({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})

filtered_df = final_upgrade[final_upgrade['Status'].isin(['DEPARTED', 'LANDED'])]

flights_per_day = final_upgrade['DayOfWeek'].value_counts().reindex([
'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])

delayed_flights_per_day = final_upgrade[final_upgrade['Total Delay (minutes)'] > 0]['DayOfWeek'].value_counts().reindex([
'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])

delay_percentage_per_day = (delayed_flights_per_day / flights_per_day) * 100

cancelled_flights_per_day = final_upgrade[final_upgrade['Status'] == 'CANCELED']['DayOfWeek'].value_counts().reindex([
'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

flights_per_day.plot(kind='bar', color='purple', ax=axes[0])
axes[0].set_title('Number of flights by day of week')
axes[0].set_xlabel('Day of week')
axes[0].set_ylabel('Number of flights')
axes[0].tick_params(axis='x', rotation=45)
for i, v in enumerate(flights_per_day):
    axes[0].text(i, v + 3, str(v), ha='center', color='black', fontweight='bold')

delay_percentage_per_day.plot(kind='bar', color='#FF8C00', ax=axes[1])
axes[1].set_title('Percentage of delayed flights by day of week')
axes[1].set_xlabel('Day of week')
axes[1].set_ylabel('Percentage of flights delayed')
axes[1].tick_params(axis='x', rotation=45)
for i, v in enumerate(delay_percentage_per_day):
    axes[1].text(i, v + 1, f'{v:.1f}%', ha='center', va='bottom', color='black', fontsize=9, fontweight='bold')

cancelled_flights_per_day.plot(kind='bar', color='salmon', ax=axes[2])
axes[2].set_title('Number of cancelled flights by day of week')
axes[2].set_xlabel('Day of week')
axes[2].set_ylabel('Number of cancelled flights')
axes[2].tick_params(axis='x', rotation=45)
for i, v in enumerate(cancelled_flights_per_day):
    axes[2].text(i, v + 1, str(v), ha='center', va='bottom', color='black', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
#Visualization 3:Top 10 countries by number of flights
#An airport can determine which airlines use its services most frequently. 
#This can help prioritize maintenance and infrastructure improvements for these airlines.

flight_counts = final_upgrade['Country'].value_counts()
top_countries = flight_counts.head(10)

df = pd.DataFrame({'Country': top_countries.index, 'Flights': top_countries.values})
df['Percentage'] = (df['Flights'] / df['Flights'].sum()) * 100

explode = [0.1 if i == df['Percentage'].idxmax() else 0 for i in range(len(df))]
palette_rocket = sb.color_palette("rocket", 10)
plt.figure(figsize=(8, 8))
plt.pie(df['Percentage'], labels=df['Country'], autopct='%1.1f%%', shadow=True, explode=explode, colors=palette_rocket, pctdistance=0.85)
plt.title('Top 10 countries by number of flights')
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('equal')
plt.show()
```

```python
#Visualization 4: Top 5 Airlanes with Maximum Cancellations During the Day
#Analyzing canceled flights by time and country can help airports determine whether additional measures 
#are needed in terminals to reduce inconvenience in the event of delays or cancellations.

canceled_flights = final_upgrade[final_upgrade['Status'] == 'CANCELED']
top_5_airlines = canceled_flights['Airline'].value_counts().nlargest(5).index
plt.figure(figsize=(14, 7))
for airline in top_5_airlines:
    data1 = canceled_flights[canceled_flights['Airline'] == airline]
    data1['HourBin'] = data1['PlanTime'].dt.hour // 2 * 2 
    data_grouped = data1.groupby('HourBin').size()

    plt.plot(data_grouped.index, data_grouped.values, marker='o', label=airline)

plt.title('Top 5 Airlines with Maximum Cancellations During the Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Cancellations')
plt.xticks(range(0, 24, 2))
plt.legend(title='Airlines')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Visualisation 5: Distribution of flight delays in minutes
#This visualization helps in understanding the distribution of flight delays.
#This can help airlines and airports in planning and managing resources efficiently.

final_upgrade['PlanTime'] = pd.to_datetime(final_upgrade['PlanTime'])
final_upgrade['ActualTime'] = pd.to_datetime(final_upgrade['ActualTime'])
final_upgrade['Delay'] = (final_upgrade['ActualTime'] - final_upgrade['PlanTime']).dt.total_seconds() / 60

plt.figure(figsize=(10, 6))
plt.hist(final_upgrade['Delay'].dropna(), bins=30, edgecolor='k')
plt.title('Distribution of flight delays')
plt.xlabel('Delay (minutes)')
plt.ylabel('Number of flights')
plt.show()
```
